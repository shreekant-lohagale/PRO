## Practical 02 : EDA and Performance against other types, height, weight classification of Pokemon dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Datasets/pokemon.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# fill the missing values with the mean of the column

df['type2'].fillna('None', inplace=True)
df['height_m'].fillna(df['height_m'].mean(), inplace=True)
df['weight_kg'].fillna(df['weight_kg'].mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
# classification of pokemons based on their type1

plt.figure(figsize=(8,6))
sns.countplot(data=df, x='type1', order=df['type1'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Number of Pokemons by Type1')
plt.show()
```

```python
# classification of pokemons based on their type2

plt.figure(figsize=(8,6))
sns.countplot(data=df, x='type2', order=df['type2'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Number of Pokemons by Type2')
plt.show()
```

```python
# scatter plot of height vs weight of pokemons

plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='height_m', y='weight_kg', hue='type1')
plt.title('Height vs Weight of Pokemons')
plt.show()
```

```python
# attack vs defense of pokemons

plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='attack', y='defense', hue='type1')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.title('Attack vs Defense of Pokemons')
plt.show()
```

```python
# attack vs defense of pikachus, balbasaur, charmander, squirtle, charizard, Lucario, Snorlax

selected_pokemons = df[df['name'].isin(['Pikachu', 'Bulbasaur', 'Charmander', 'Squirtle', 'Charizard', 'Lucario', 'Snorlax'])]
plt.figure(figsize=(10,6))
sns.scatterplot(data=selected_pokemons, x='attack', y='defense', hue='name')
plt.title('Attack vs Defense of Pikachu, Bulbasaur, Charmander, Squirtle, Charizard, Lucario, Snorlax')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid()
plt.show()
```

```python
# battle between two pokemons based on their total stats
# Verify base_total is sum of individual stats
# df[['hp', 'attack', 'defense', 'sp_attack', 'sp_defense', 'speed', 'base_total']].head()

def battle(pokemon1, pokemon2):
    p1 = df[df['name'] == pokemon1].iloc[0]
    p2 = df[df['name'] == pokemon2].iloc[0]
    
    print(f"Battle between {pokemon1} and {pokemon2}\n")
    print(f"{pokemon1} Total Stats: {p1['base_total']}")
    print(f"{pokemon2} Total Stats: {p2['base_total']}")
    
    if p1['base_total'] > p2['base_total']:
        print(f"\n{pokemon1} wins!")
    elif p1['base_total'] < p2['base_total']:
        print(f"\n{pokemon2} wins!")
    else:
        print("\nIt's a tie!")
battle('Pikachu', 'Charizard')
```

```python
import random

pokemon_list = df['name'].tolist()

pokemon1 = random.choice(pokemon_list)
pokemon2 = random.choice(pokemon_list)

battle(pokemon1, pokemon2)
```
